### House Rent Prediction Dataset

##### El siguiente Data Frame posee informacion de mas de 4700 casas, apartamentos y flats en renta actualmente en India. Se presentan diferentes parametros como el tamaño, numero de pisos, tipo de area, localidad, ciudad, estado del mobiliario, numero de baños, asi como su precio de renta.

```python
house_rent = pd.read_csv('/workspaces/data-science/data/raw/House_Rent_Dataset.csv')
print(house_rent.shape)
house_rent.info()
house_rent.head()
```

##### Multiples variables de este dataset poseen un gran valor a la hora de realizar analisis que relacionen las caracteristicas del inmueble y su precio. Sera posible encontrar cuales parametros son considerados en India como un factor capaz de aumentar el precio de renta, como por ejemplo su ubicacion.

```python
renta_promedio = house_rent.groupby("Area Locality")["Rent"].mean()
ubicacion_max_renta = renta_promedio.idxmax()
print("Ubicacion con renta mas alta: ",ubicacion_max_renta)
```

### Restaurant's cuisine ratings

##### El siguiente dataset contiene informacion de la situacion sociocultural y habitos de consumo de alcohol y tabaco de una muestra de personas y su relacion con la puntuacion que otorgaron a platos de diferentes categorias.

```python
cuisine = pd.read_csv('/workspaces/data-science/data/raw/Cuisine_rating.csv')
print(cuisine.shape)
cuisine.info()
cuisine.head()
```

##### Existe un gran abanico de posibilidades de analisis de datos en este data frame. Algunos ejemplos:

##### La relacion entre la preferencia a determinados tipos de comidas y el estado sociocultural de la persona, como su nivel socioeconomico o estado civil.
##### La comparacion entre los puntajes otorgados a diferentes platos por personas de similares caracteristicas socioculturales segun si son fumadores o no.
##### Trends de puntajes a platos segun genero.

```python
fila1 = cuisine.loc[(cuisine['Gender'] == "Male") & (cuisine['Marital Status'] == "Single") & (cuisine['Activity'] == "Student") & (cuisine["Smoker"] == "Never")]

fila2 = cuisine.loc[(cuisine['Gender'] == "Male") & (cuisine['Marital Status'] == "Single") & (cuisine['Activity'] == "Student") & (cuisine["Smoker"] == "Often")]

counts1 = fila1["Cuisines"].value_counts()
counts2 = fila2["Cuisines"].value_counts()

print("Preferencia de cuisine entre hombres solteros estudiantes: \n",counts1)
print("Preferencia de cuisine entre hombres solteros estudiantes fumadores: \n",counts2)
```

### Cause of Deaths around the World

##### El siguiente data frame contiene informacion sobre las causas de muertes de la poblacion mundial en los años 1990 hasta 2019. Recopila variables como pais o territorio, asi como una gran variedad de causas de defuncion registradas.

```python
cause_death = pd.read_csv('/workspaces/data-science/data/raw/cause_of_deaths.csv')
print(cause_death.shape)
cause_death.info()
cause_death.head()
```

##### Los datos presentes en este data frame otorgan una muestra confiable y representativa de la ubicacion, año, causas y numero de muertes en el mundo. Estos datos tienen un exelente potencial para el desarrollo de un analisis sobre la manera en la que las causas de muerte prevalentes fluctuan a lo largo de los años y su relacion con una comparativa entre paises desarrollados o en vias de desarrollo. Otra tematica de analisis potencial que permiten los datos es la influencia de aspectos culturales o propios de la situacion geografica sobre la prevalencia en causas de defuncion.

```python
pais = "Argentina"
año = 2019

fila = cause_death.loc[(cause_death['Country/Territory'] == pais) & (cause_death['Year'] == año)]
columna_max = fila.select_dtypes(include=['int64', 'float64']).idxmax(axis=1)
print(columna_max)
```
